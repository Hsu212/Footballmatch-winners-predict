Cleaning the data for ML

```python
matches.dtypes #check the data types
```

```python
matches["date"] = pd.to_datetime(matches["date"]) #Change the date from object to datetime
matches.dtypes
```

Create the predictions for ML

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes #.astype("category").cat.codes is to chg String into category and into numbers
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
matches
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex = True).astype("int") #.str.replace(":.+", "", regex = True)is method to replace the minute into nth
#.astype("int") is to chg the hour into int type
matches
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
matches
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
matches
```

Create the initial ML model

```python
from sklearn.ensemble import RandomForestClassifier #RandomForestClassifier is to pick up non-linearities
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1) #the higher n_estimators the more runtime but get the more accurate values
#min_samples_splt is number of samples in the leaf in decision trees before splitting the node,the higher it is, the less overfit and lower accuracy
#random_state means that if we run multiple times,will get the same result
```

```python
train = matches[matches["date"] < '2022-01-01']
test = matches[matches["date"] > '2022-01-01']

predictors = ["venue_code","opp_code","hour","day_code"]

rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["target"], preds)
acc
```

```python
combined = pd.DataFrame(dict(actual = test["target"], prediction = preds))

pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
from sklearn.metrics import precision_score
precision_score(test["target"], preds)
```

Improving the precision with rolling averages

```python
grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Manchester City") #This line is to look at a single group
group
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values("date")
  rolling_stats = group[cols].rolling(3, closed='left').mean() #to watch the datas for first 3 weeks and closed='left' is to take the current week out
  group[new_cols] = rolling_stats
  group = group.dropna(subset = new_cols)
  return group
```

```python
cols = ["gf","ga","sh","sot","dist","fk","pk","pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

Retraining ML model

```python
def make_predictions(data, predictors):
  train = data[data["date"] < '2022-01-01']
  test = data[data["date"] > '2022-01-01']
  rf.fit(train[predictors], train["target"])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual = test["target"], predicted = preds), index = test.index)
  precision = precision_score(test["target"], preds)
  return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
precision
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index = True, right_index = True)
combined
```

Combining home and away predictions

```python
class MissingDict(dict):
  _missing_ = lambda self, key: key

map_values = {
    "Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"
}
mapping = MissingDict(**map_values)
```

```python
mapping["Brighton and Hove Albion"]
```

```python
combined["new_team"] = combined["team"].map(mapping)
combined
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"]) #to see predictions match up on both sides
merged
```
